```python
import matplotlib.pyplot as plt 
import random as rd
import numpy as np 
import sklearn.datasets as skl
%matplotlib inline
```

Learning Colors

```python
x=np.linspace(0,20,100)
y=np.sin(x)

plt.plot(x,y,c='tomato')
```

```python
print('Demo for colour cycle')
for offset in np.arange(10):
    plt.plot(x,y+offset)
```

```python
print('Demo for Specified colour cycle')
for offset,colour in zip(np.arange(10),list('rgbcmykw')):
    plt.plot(x,y+offset,c=colour)
```
